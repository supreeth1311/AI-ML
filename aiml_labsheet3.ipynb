```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('Year')
plt.ylabel('Yield')
plt.show()
```

```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('crop yiekt in kanto')
plt.legend(['Apple','Orange'])
plt.show()
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('crop yield in kanto')
plt.legend(['Apple','Orange'])
plt.show()
```

```python
plt.plot(year,apples,marker='s',c='b',ls='-',lw='2',ms=8,mew=2,mec='navy')
plt.plot(year,oranges,marker='x',ls='--',lw='3',alpha=0.5)
plt.xlabel('Year')
plt.ylabel('Yield')
plt.title('crop yields in kanto')
plt.legend(['Apple','Orange'])
plt.show()
```

```python
plt.plot(year,apples,'s-b')
plt.plot(year,oranges,'o-r')
plt.xlabel('Year')
plt.ylabel('Yield')
plt.legend(['Apple','Orange'])
plt.show()
```
